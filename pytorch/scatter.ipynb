## scatter(dim, index, src)

```python
# 1-d case: scatter label into a larger vector
label = torch.randint(184, (8,))
print(label)
b = torch.zeros(184,)
b.scatter_(0, label, 1.0)
```

```python
# 2-d case
label = torch.randint(184, (8,2))
print(label)
b = torch.zeros(184,2)
b.scatter_(0, label, 1.0)# scatter value alone the dim. dim 0
```

```python
# exercise:
# objs: (O,) values in [0, num_classes)
# obj_to_img: (O,) like: (00000,1111,22,333) lenght=O, 
# output: N*num_classes. Multi-hot matrix.
num_cls = 15
O = 20
objs = torch.randint(num_cls, (O,))
obj_to_img = [0]*5+[1]*4+[2]*2+[3]*3+[4]*(O-14)
print(objs)
print(obj_to_img)
```

```python
from collections import Counter

split_idx = list(Counter(obj_to_img).values())
print(split_idx)
objs_list = objs.split(split_idx)
print(objs_list)
```

```python
multi_hot = []
for label in objs_list:
    tmp = torch.zeros(1, num_cls)
    tmp.scatter_(1, label.unsqueeze(0), 1.0)
    multi_hot.append(tmp)
multi_hot = torch.cat(multi_hot, dim=0)
print(multi_hot)
```
