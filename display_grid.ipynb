```python
import matplotlib.pyplot as plt 
import cv2
from glob import glob
```

```python
mono1_list = glob("samples/mono_*/*.jpeg")
mono1_list.sort()
stereo_list = glob("samples/stereo*/*.jpeg")
stereo_list.sort()
real_list = glob("samples/real/*.jpg")
real_list.sort()
mono2_list  = glob("samples/mono+*/*.jpeg")
mono2_list.sort()
print(mono1_list[:5])
print(stereo_list[:5])
print(real_list[:5])
print(mono2_list[:5])
```

```python
import numpy as np 
import random as rnd 
from PIL import Image

real = Image.open(real_list[0])
mono = Image.open(mono1_list[0])
mono2 = Image.open(mono2_list[0])
stereo = Image.open(stereo_list[0])
print(real.size, real.mode)
print(mono.size, mono.mode)
print(mono2.size, mono2.mode)
print(stereo.size, stereo.mode)
real_np = np.array(real)
mono_np = np.array(mono)
mono2_np = np.array(mono2)
stereo_np = np.array(stereo)
print(real_np.shape)
print(mono_np.shape)
print(mono2_np.shape)
print(stereo_np.shape)
```

```python
## image grid.
from mpl_toolkits.axes_grid1 import ImageGrid

fig = plt.figure(figsize=(18,8))
grid = ImageGrid(fig, 111, nrows_ncols=(1,4), axes_pad=0.1)
for ax, im in zip(grid, [real_np, mono_np, mono2_np, stereo_np]):
    ax.imshow(im)

plt.show()
```
